## plot of the demand curve, which is transformed to bitcode.

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = [10,6] # graph size
mpl.rcParams['lines.linewidth'] = 3  # line width
mpl.rcParams['lines.markersize'] = 10  # size of point markers
mpl.rcParams['xtick.labelsize'] = 14  # font size of x axis ticks
mpl.rcParams['ytick.labelsize'] = 14 # font size of y axis ticks
mpl.rcParams['axes.labelsize'] = 'larger'  # font size of the axes labels
```

```python
x = np.linspace(0, 100, 101)
a = -1/100
b = 1

plt.plot(x, a*x+b, lw=3)
plt.axhline(0, -5, 5, color='k', lw=0.5)
plt.axvline(0, -5, 5, color='k', lw=0.5)
plt.xlabel('Price per unit [money]', fontsize=15)
plt.ylabel('Demand for one unit [fraction]', fontsize=15)
plt.title('Demand curve for a lemonade', fontsize=18)
plt.savefig('demand_lim.png', dpi=300)
plt.show()
```

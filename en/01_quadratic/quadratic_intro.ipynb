# Quadratic function

## General form of a quadratic function is:
 
$ f(y)=ax^2+bx+c $

where *a,b,c* are parameters (constants) and functinal values (y axis) are defined for every real value x (x axis)
 
---
In practice, quadratic function/equation is useful in following situations:

* Estimates of material consumption for enlarging objects/areas etc.
* Calculation of optimal price at which to sell a product.
* Physics of accelarating motion kinetics, mechanics of tossing a ball etc.

## Apex
* Every quadr. function has an apes (which can be its minimum mor maximum)
* We try to find this extreme value for example in 'Selling limonade' problem

You might remember that apex x value is:

$x=-\dfrac{b}{2a}$

Let's confirm that graphically

```python
# x_apex is (value on the x axis)
x_apex = -b/(2*a)

# We calculate functional value as above,
# now only for one value instead of 101.
fy_apex = a*x_apex**2 + b*x_apex + c
```

```python
# Plot the APEX point [x_apex, fy_apex] in addition to our function
fig = plt.figure(figsize=(10,6))
plt.plot(x, fy, lw=3)
plt.plot(x_apex, fy_apex, 'ro', ms=10) # 'r' plots in red and 'o' means circular marker.

# adding lines for x and y axes
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='k', lw=0.5)

# zoom to the apex by selecting limits of x axis (xlim) and y axis (ylim)
plt.xlim([-1, 4])
plt.ylim([-2.5,4])
plt.show()
```

## Intercept with x axis
* Not every quadratic function intercepts x axis.
* It can have 0,1 or 2 intercepts.

Number of intercepts is determined by the value of discriminant D:

$D=b^2-4ac$

Intercepts are then in x values x1 and x2:

$x_{1,2}=\dfrac{-b\pm\sqrt{D}}{2a}$

Again, let's confirm that graphically

```python
## function 1
a = 2
b = -5
c = 1
# functional values 1
f1y = a*x**2 + b*x + c

# discriminant 1
D1 = b**2-4*a*c
```

```python
# function 2
a = 2
b = 0
c = 1

f2y = a*x**2 + b*x + c

D2 = b**2-4*a*c
```

```python
# function 3
a = 1
b = 2
c = 1

f3y = a*x**2 + b*x + c

D3 = b**2-4*a*c
```

```python
# Plot all three functions
fig = plt.figure(figsize=(10,6))
plt.plot(x, f1y, 'red', lw=3)
plt.plot(x, f2y, 'blue', lw=3)
plt.plot(x, f3y, 'green', lw=3)

# add axes
plt.axhline(0, -5, 5, color='k', lw=0.5)
plt.axvline(0, -5, 5, color='k', lw=0.5)

# We can zoom in our plot by limiting (xlim, ylim) range of the plot.
plt.xlim([-5,5])
plt.ylim([-3,10])
plt.show()
```

* Red one (f1y) has 0 intercepts
* Blue (f2y) has 2
* Green (f3y) has 1 

## We can print values of discriminants to confirm the textbook knowledge

```python
# function print can be used in several ways to print values stored within variables
print(f'Discriminant for f1y: {D1}')
print('Discriminant for f2y:', D2)
print('Discriminant for f3y: {0}'.format(D3))
```

## Yes, 
* D>0, two intercepts
* D=0, 1 intercept
* D<0, 0 intercepts

## One of the more advanced concepts is focus of the quadratic function
* Every quadratic function has one
* Focus is significant for example for imiging in telescopes or concentrating sun-rays in solar panel electricity generation / or cooking.

Coordinated of the focus are:

$F=\left[ -\dfrac{b}{2a}, \dfrac{4ac-b^2+1}{4a} \right]$

If we compare focus coordinates with those of the apex we calculated above:

$V=\left[ -\dfrac{b}{2a}, \dfrac{4ac-b^2}{4a} \right]$

1. X coordinate is identical, so focus lies on the symmetry axis of the parabola.
2. Y coordinate is vertically shifted by $\dfrac{1}{4a}$, 
3. This value of shift is called focal length. Note that it contains only parameter *a*, which is responsiblefor curvature of the quadratic function.

## Let's plot all points we mentioned above
* Using function 1 from above

```python
# function 1
a = 2
b = -5
c = 1

f1y = a*x**2 + b*x + c

# Discriminant
D1 = b**2-4*a*c

incept_x = np.array([ (-b + np.sqrt(D1))/(2*a),
                      (-b - np.sqrt(D1))/(2*a),
                    ])
incept_y = a*incept_x**2 + b*incept_x + c

# Apex
V = [-b/(2*a), (4*a*c-b**2)/(4*a)]
# Focus
F = [-b/(2*a), (4*a*c-b**2+1)/(4*a)]
```

```python
incept_x
```

```python
# Plot it all
fig, ax = plt.subplots(figsize=(10,6))
ax.set_aspect(1) # set the same aspect ratio for x and y axes
plt.plot(x, f1y, lw=3)  # parabola
plt.plot(incept_x, incept_y, 'go', ms=10)  # intercepts with x (solution of  quadr. equation with right side = 0)
plt.plot(V[0], V[1], 'ro', ms=10)  # apex
plt.plot(F[0], F[1], 'ko', ms=10)  # focus

# add axes
plt.axhline(0, -5, 5, color='k', lw=0.5)
plt.axvline(0, -5, 5, color='k', lw=0.5)

# limit graph range
plt.xlim([-0.2,2.7])
plt.ylim([-2.5,1])
plt.show()
```

## Vrchol
* Každá kvadratická funkce má vrchol (minimum či maximum)
* Tento bod hledáme například v naší *podnikatelské* úloze

Ze školy si možná pamatujete, že vrchol je v bodě

$x=-\dfrac{b}{2a}$

Tak si to oveřme graficky

```python
# x_vrchol je
x_vrchol = -b/(2*a)

# funkční hodnotu spočítáme stejně jako výše,
# teď je ale vstupem jen jedna hodnota x, a ne 101
fy_vrchol = a*x_vrchol**2 + b*x_vrchol + c
```

```python
# k původnímu grafu vykreslíme náš jeden bod [x_vrchol, fy_vrchol]
fig = plt.figure(figsize=(10,6))
plt.plot(x, fy, lw=3)
plt.plot(x_vrchol, fy_vrchol, 'ro', ms=10) # r znamená vykreslí červeně (od 'red') a o značí symbol kolečka

plt.xlim([-2, 1.5])
plt.ylim([0,4])
plt.show()
```

## Průsečík s osou x
* NE každá kvadratická funkce protína osu x
* Tedy, může mít 0,1 nebo 2 takzvané průsečíky.

Možná si pamatuješ, že počet průsečíků se určí odmocninou z diskriminantu

$\sqrt{D}=\sqrt{b^2-4ac}$

Průsečíky jsou pak na hodnotách x1, x2

$x_{1,2}=\dfrac{-b\pm\sqrt{b^2-4ac}}{2a}$

Tak si to vše oveřme opět graficky

```python
## funkce 1
a = 2
b = 0
c = 1

# funkční hodnoty
f1y = a*x**2 + b*x + c

# diskriminant
D1 = b**2-4*a*c
```

```python
# funkce 2
a = 2
b = -5
c = 1

f2y = a*x**2 + b*x + c

D2 = b**2-4*a*c
```

```python
# funkce 3
a = 1
b = 2
c = 1

f3y = a*x**2 + b*x + c

D3 = b**2-4*a*c
```

```python
# Opět vše vykreslíme
fig = plt.figure(figsize=(10,6))
plt.plot(x, f1y, 'red', lw=3)
plt.plot(x, f2y, 'blue', lw=3)
plt.plot(x, f3y, 'green', lw=3)

plt.axhline(0, -5, 5, color='k', lw=0.5)
plt.axvline(0, -5, 5, color='k', lw=0.5)

# přiblížíme se počátku díky limitě rozsahu našeho x a y
plt.xlim([-5,5])
plt.ylim([-3,10])
plt.show()
```

* vydíme, že červená (f1y) má 0 průsečíků
* modrá (f2y) má 2
* zelená (f3y) má jeden

## Jedním z pokročilejších konceptů je ohnisko neboli fokus paraboly/kvadratické funkce
* každá kv. funkce má ohnisko.
* Je vyznamné například pro zobrazování dalekohledy, ale i koncentraci sluneční energie pro vyrobu el. energie či vaření.

výrazy pro souřadnice ohniska jsou:

$F=\left[ -\dfrac{b}{2a}, \dfrac{4ac-b^2+1}{4a} \right]$

pokud si porovnáme ohnisko se souřadnicemi vrcholu, které jsme počítali výše:

$V=\left[ -\dfrac{b}{2a}, \dfrac{4ac-b^2}{4a} \right]$

1. Souřadnice x je totožná, tedy ohnisko leží na ose symetrie paraboly.
2. Souřadnice y je jen posunuta vertikálně o $\dfrac{1}{4a}$, 
3. Této hodnotě se říká ohnisková vzdálenost, a všimněte si že záleží jen na *a*, které určuje jak moc se parabola zakřivuje.

## Pojďme vykreslit vše, co jsme dopusud probrali
* k tomu vezměme funkci 2 z výše

```python
# funkce 2
a = 2
b = -5
c = 1

f2y = a*x**2 + b*x + c

# Diskriminant
D2 = b**2-4*a*c

reseni_x = np.array([ (-b + np.sqrt(D2))/(2*a),
                      (-b - np.sqrt(D2))/(2*a),
                    ])
reseni_y = a*reseni_x**2 + b*reseni_x + c

# Vrchol
V = [-b/(2*a), (4*a*c-b**2)/(4*a)]
# Ohnisko
F = [-b/(2*a), (4*a*c-b**2+1)/(4*a)]
```

```python
reseni_x
```

```python
# Opět vše vykreslíme
fig = plt.figure(figsize=(10,6))
plt.plot(x, f2y, lw=3)  # parabola
plt.plot(reseni_x, reseni_y, 'go', ms=10)  # průsečiky s x (řešení kv. rovnice s pravou stranou = 0)
plt.plot(V[0], V[1], 'ro', ms=10)  # vrchol
plt.plot(F[0], F[1], 'ko', ms=10)  # ohnisko

plt.axhline(0, -5, 5, color='k', lw=0.5)
plt.axvline(0, -5, 5, color='k', lw=0.5)

# limita rozsahu grafu
plt.xlim([-0,3])
plt.ylim([-2.5,1])
plt.show()
```
